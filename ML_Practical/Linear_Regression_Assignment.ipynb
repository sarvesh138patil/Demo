```python
#import dependencies
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.model_selection import train_test_split
```

```python
#Load the Boston Housing Data set from sklearn.datasets and print it
from sklearn.datasets import load_boston
boston = load_boston()
#print(boston)
```

```python
#transform the data set into a data frame
# data = independent variable (x values)
# feature names = column names of the data
# target = target variable or price or dependent variable (y values)

df_x = pd.DataFrame(boston.data, columns=boston.feature_names)
df_y = pd.DataFrame(boston.target)
```

```python
#get some statistics from data set 
df_x.describe()
```

```python
#Initialize a linear regression model
reg = linear_model.LinearRegression()
```

```python
#Split the data into 67% training and 33% testing data
#We have to split the independent variables (x) and dependen variables (y)
x_train, x_test, y_train, y_test = train_test_split (df_x, df_y, test_size = 0.33, random_state = 42)

#Train our model with training data
reg.fit(x_train, y_train)
```

```python
#Print the coefficients for each feature/column of our model 
print(reg.coef_)
```

```python
#Print our price prediction on our test data
y_pred = reg.predict(x_test)
print(y_pred)
```

```python
#Print the prediction for the third row of our test data actual price = 13.6
y_pred[2]

#Print the actual price of houses from the testing data set
y_test[0]
```

```python
#Two different ways to check model performance/accuracy using,
#mean squared error which tells you how close a regression line is to a set of points

#Mean squared error by numpy
print(np.mean((y_pred-y_test)**2))

#Mean squared error by sklearn 
from sklearn.metrics import mean_squared_error
print(mean_squared_error(y_test,y_test))
```
